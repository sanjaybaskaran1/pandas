```python
makes_list = df['make'].unique().tolist()
print(makes_list)
```

```python
Car_price=[]
for make in makes_list:
    print("car model name=",make)
    price=make_grp.get_group(make)['price'].sum()
    print("car count=",make_grp.get_group(make)['price'].count())
    print("Total price =",price)
    Car_price.append((make,price))
    print('-----------------------------------------------------------------------------------')
```

```python
make_fuel_grp = df.groupby(by=["make","fuel-type"])
make_fuel_grp
```

```python
Car_Type = []
for name,grp in make_fuel_grp:
    Car_Type.append(name)
    print(name)
```

```python
make_fuel_grp.get_group(('mercedes-benz', 'diesel'))["price"]
```

```python
make_fuel_grp.get_group(('mercedes-benz', 'diesel'))["price"].sum()
```

```python
for i in Car_Type:
    print("Group =",i)
    print("Car Count =",make_fuel_grp.get_group(i)["price"].count())
    print("Total Price =",make_fuel_grp.get_group(i)["price"].sum())
    print("Maximum Price",make_fuel_grp.get_group(i)["price"].max())
    print("Minimum Price =",make_fuel_grp.get_group(i)["price"].min())
    print("Average Price =",make_fuel_grp.get_group(i)["price"].mean())
    print("******************************************")
```
