```python
#change the datatype of hire_date from(object) to (datetime) inorder to performe time series analysis
df['hire_date']=pd.to_datetime(df['hire_date'])
df['hire_date']
```

```python
#find the year of all the rows using {hire_date}
#and add that new_column-->{year} to our existing data frame
df['year']=df['hire_date'].dt.year
df['year']
```

```python
#find the quarter of all the rows using {hire_date}
#and add that new_column-->{quarter} to our existing data frame
df['quarter']=df['hire_date'].dt.quarter
df['quarter']
```

```python
#find the month_name of all the rows using {hire_date}
#and add that new_column-->{month_name} to our existing data frame
df['month_name']=df['hire_date'].dt.month_name()
df['month_name']
```

```python
#find the day_name of all the rows using {hire_date}
#and add that new_column-->{day_name} to our existing data frame
df['day_name']=df['hire_date'].dt.day_name()
df['day_name']
```

```python
#create a list called {months_list} and store all the {month names} inside that list
months_list = df['month_name'].unique().tolist()
months_list
```

```python
#find the totalm salary count, total salary, max, min, avg salary 
#for the employees in the year 1990 & (for all the months in that year) 
for month in months_list:
    filt = (df['year2'] == 1990) & (df['month_name'] == month)
    print("month name=", month,":")
    print('**********************')
    print("total salary counts =",df.loc[filt]["salary"].count())
    print("total salary =",df.loc[filt]["salary"].sum())
    print("max salary =",df.loc[filt]["salary"].max())
    print("min salary =",df.loc[filt]["salary"].min())
    print("avg salary =",df.loc[filt]["salary"].mean())
    print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
    print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
```
