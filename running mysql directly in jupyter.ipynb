```python
%%sql
SELECT 
    e.emp_no, e.first_name, e.last_name, d.dept_name
FROM 
    employees e
JOIN 
    dept_manager dm ON e.emp_no = dm.emp_no
JOIN 
    departments d ON d.dept_no = dm.dept_no;
```

```python
result = %%sql
         SELECT 
         e.emp_no, e.first_name, e.last_name, d.dept_name
         FROM 
         employees e
         JOIN 
         dept_manager dm ON e.emp_no = dm.emp_no
         JOIN 
         departments d ON d.dept_no = dm.dept_no;
```

```python
result = %sql SELECT e.emp_no, e.first_name, e.last_name, d.dept_name 
              FROM employees e 
              JOIN dept_manager dm ON e.emp_no = dm.emp_no 
              JOIN departments d ON d.dept_no = dm.dept_no;
```

```python
result = %sql SELECT e.emp_no, e.first_name, e.last_name, d.dept_name \
              FROM employees e \
              JOIN dept_manager dm ON e.emp_no = dm.emp_no \
              JOIN departments d ON d.dept_no = dm.dept_no;
```

```python
employees_df = pd.DataFrame(result)
```
