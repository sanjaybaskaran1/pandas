```python
#updating rows in pandas
student = {
    "student_name":["lavanya","jaanu","lakshith","dakshith"],
    "student_degree":["bca","bca","third","fifth"],
    "student_grade":["A","A","B","C"]
}
print(student)
```

```python
import pandas as pd
df = pd.DataFrame(student)
df
```

```python
# 1) updating rows using concat() method
# New data to add
new_data = pd.DataFrame([{"student_name": "aakash", "student_degree": "engineering", "student_grade": "A++"}])

# Concatenate the original DataFrame with the new data
df2 = pd.concat([df, new_data], ignore_index=True)

print(df2)
```
